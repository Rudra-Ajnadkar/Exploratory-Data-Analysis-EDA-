```python
# Data Cleaning
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="Pclass",y="Age",data=df)
```

```python
def input_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        if Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df["Age"]=df[["Age","Pclass"]].apply(input_age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop("Cabin",axis=1,inplace=True)
```
